```python
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import json
```

```python
rsp = gpd.read_file("rsp-data/rsp-3.geojson")
```

```python
rsp.head()
```

```python
rsp.columns
```

```python
weights = {'HISTO_forest': 7, 'HISTO_mangrove': 6, 'HISTO_grassland': 5, 'HISTO_crops': 3, 'HISTO_bare_land': 2, 'HISTO_settlements': 1, 'HISTO_water': 0}

rsp['RAS'] = sum(rsp[col] * w for col, w in weights.items()) / rsp[list(weights.keys())].sum(axis=1)
```

```python
# STEP 1 - Shapiro-Wilkes test for normality
```

```python
# Perform Shapiro-Wilk Test
stat_temp, p = stats.shapiro(rsp['Temp (C)'])
```

```python
print(f"Statistic: {stat_temp:.4f}, p-value: {p:.4f}")
if p > 0.05:
    print("Data looks Gaussian (fail to reject H0)")
else:
    print("Data does not look Gaussian (reject H0)")
```

```python
import geopandas as gpd
import pandas as pd
from scipy import stats

# rsp is already a GeoDataFrame
gdf = rsp

# Get the numeric columns
numeric_cols = ['Temp (C)', 'Baro (mb)', 'Depth (m)', 'pH', 'ORP (REDOX)']
numeric_cols = [col for col in numeric_cols if col in gdf.columns]

shapiro_results = []

# Loop through each numeric column
for col in numeric_cols:
    # Remove NaN values - use pd.to_numeric, not rsp.to_numeric
    data = pd.to_numeric(gdf[col], errors='coerce').dropna()
    
    # Perform Shapiro-Wilks test
    if len(data) > 3:  # Shapiro-Wilks requires at least 3 samples
        statistic, p_value = stats.shapiro(data)
        
        shapiro_results.append({
            'Column': col,
            'Statistic': statistic,
            'P-Value': p_value,
            'Sample_Count': len(data),
            'Is_Normal': 'Yes' if p_value > 0.05 else 'No',
            'Mean': data.mean(),
            'Std_Dev': data.std(),
            'Min': data.min(),
            'Max': data.max()
        })

# Create results dataframe
results_df = pd.DataFrame(shapiro_results)

print(results_df)

# Save to CSV if desired
results_df.to_csv('shapiro_wilks_results.csv', index=False)
```

```python
import geopandas as gpd
import pandas as pd
from scipy import stats

# rsp is already a GeoDataFrame
gdf = rsp

# Get the numeric columns
numeric_cols = ['Temp (C)', 'Baro (mb)', 'Depth (m)',
       'pH', 'ORP (REDOX)', 'DO (% Sat)', 'DO (mg/L)', 'EC (uS/cm @25C)',
       'RES (Ohms.cm)', 'TDS (mg/L)', 'SAL (PSU)', 'SSG (st)',
       'Turbidity (NTU)', 'Nitrate (mg/L)', 'Lat', 'Lon', 'Alt (m)']
numeric_cols = [col for col in numeric_cols if col in gdf.columns]

shapiro_results = []

# Loop through each numeric column
for col in numeric_cols:
    # Remove NaN values - use pd.to_numeric, not rsp.to_numeric
    data = pd.to_numeric(gdf[col], errors='coerce').dropna()
    
    # Perform Shapiro-Wilks test
    if len(data) > 3:  # Shapiro-Wilks requires at least 3 samples
        statistic, p_value = stats.shapiro(data)
        
        shapiro_results.append({
            'Column': col,
            'Statistic': statistic,
            'P-Value': p_value,
            'Sample_Count': len(data),
            'Is_Normal': 'Yes' if p_value > 0.05 else 'No',
            'Mean': data.mean(),
            'Std_Dev': data.std(),
            'Min': data.min(),
            'Max': data.max()
        })

# Create results dataframe
results_df = pd.DataFrame(shapiro_results)

print(results_df)

# Save to CSV if desired
results_df.to_csv('shapiro_wilk_results.csv', index=False)
```
